## Object Oriented API

```python
fig, ax = plt.subplots(2)


ax[0].plot(n1, sin(n1),'r-.')
ax[1].plot(n1, cos(n1))
plt.show()
```

```python
fig = plt.figure()

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x2, y2, 'r')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title');
```

## Error Bar Chart

```python
j4 = arange(0,4,0.2)

j5 = np.exp(-j4)

error1 = 0.1 * abs(np.random.randn(len(j4)))

plt.errorbar(j4,j5 ,yerr= error1,fmt='.2')

plt.title('Error plot')
```

## Stacked Bar Chart

```python
A = [12., 30., 40., 72.]

B = [10., 20., 10., 20.]

z1 = range(4)

plt.bar(z1, A)
plt.bar(z1,B,bottom=A)
```
